```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from setfit import SetFitModel, SetFitTrainer
from datasets import Dataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 df = pd.read_csv('job_classification.csv').drop(columns='Unnamed: 0').rename(columns={'class_id': 'label'}).dropna()
```

```python
df.isna().sum()
```

```python
#is it balanced
df['label'].value_counts(normalize=True)*100
```

```python
# there are 6 labels we can use accuracy metric
df['description'].str.split().map(len).describe()
```

# WE use SefFit which is a transformer based texct feautred classification

## We concatenate title and description so the model will look at is as whole:

```python
df['text'] = df['title'] + '\n' + df['description']
```

```python
np.random.seed(22)
train_df, test_df = train_test_split(df, test_size = .2, stratify=df['label'])
train_df['label'].value_counts(normalize=True)
len(train_df), len(test_df)
```

# SetFit

## Since it is a drill and we do not have time, we will train only subsample of data

```python
#train_subset, _ = train_test_split(train_df, train_size=200,stratify = train_df['label'])

train_ds = Dataset.from_pandas(train_df)
test_ds = Dataset.from_pandas(test_df)
```

```python
# Load SetFit model from Hub
model = SetFitModel.from_pretrained("sentence-transformers/paraphrase-MiniLM-L6-v2")

# Create trainer
trainer = SetFitTrainer(
    model=model,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    batch_size=16,
    num_iterations=20, # Number of text pairs to generate for contrastive learning
    column_mapping={"text": "text", "label": "label"},
    num_epochs=1 # Number of epochs to use for contrastive learning
)
```

```python
trainer.train()
```

```python
train_df.columns
```

```python
trainer.evaluate()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
test_pred = trainer.model.predict(test_df.reset_index()['text'])
confusion_matrix(test_df['label'], test_pred.reshape(-1,1))
```

```python
sns.heatmap(confusion_matrix(test_df['label'], test_pred), annot=True, square=True)
plt.xlabel('pred')
plt.ylabel('GT')
```

```python
kws = set(sum(chi2_df.apply(list, axis=1),[]))
len(kws)
```

```python
chosen_kws = train_df['text'].apply(lambda text: pd.Series({kw: kw in text for kw in kws}))
chosen_kws = [col for col in chosen_kws.columns if chosen_kws[col].sum()>=50]
len(chosen_kws)
```

```python
train_df_kws = pd.concat([train_df, train_df['text'].apply(lambda text: pd.Series({f'is_{kw}': kw in text for kw in chosen_kws}))], axis=1)
test_df_kws = pd.concat([test_df, test_df['text'].apply(lambda text: pd.Series({f'is_{kw}': kw in text for kw in chosen_kws}))], axis=1)
```

```python
print(train_df_kws.shape, test_df_kws.shape)
```

#So far we prepared the unique words in texts as features to apply to LR, now we need the embedding and we will use both for LR

```python
import torch
from tqdm.auto import tqdm 
tqdm.pandas()
def extract_embeddings(text):
     tokens = trainer.model.model_body.tokenizer(text, return_tensors='pt', truncation=True).to(device='cuda')
     return trainer.model.model_body(tokens)['sentence_embedding'].to(device='cpu').numpy()

text = 'hello world'
with torch.no_grad():
    for df in [train_df_kws, test_df_kws]:
        df['emb'] = df['text'].progress_apply(extract_embeddings)


   
```

```python
for df in [train_df_kws, test_df_kws]:
    df['emb'] = df['emb'].map(lambda x: x[0])
```

```python
train_df_kws.T
```

```python
one_hot_df_train = train_df_kws[[col for col in train_df_kws.columns if 'is_' in col]]
train_features = np.concatenate([one_hot_df_train.values, np.stack(train_df_kws['emb'])],
                          axis=1)
one_hot_df_test = test_df_kws[[col for col in test_df_kws.columns if 'is_' in col]]
test_features = np.concatenate([one_hot_df_test.values, np.stack(test_df_kws['emb'])],
                          axis=1)

print(train_features.shape, test_features.shape)
```

```python
pd.DataFrame(train_features).describe()
# We see that both features type do not talk same language, chi word feature (exists or not) are between 0 to 1
# while embedding lyes in different scale so we need to scale the df
```

```python
from sklearn.preprocessing import StandardScaler

def preprocessing(features, ss=None):
    if ss == None:
        ss = StandardScaler()
        ss.fit(features)
    return ss.transform(features), ss
train_features_norm, ss = preprocessing(train_features)
test_features_norm, _ = preprocessing(test_features, ss=ss)
train_features_norm.shape  
```

```python
from sklearn.linear_model import LogisticRegression as lr
model = lr()
model.fit(train_features_norm,train_df['label'])
```

```python
pred = model.predict(test_features_norm)
print('Accuracy: ', (np.mean(pred==test_df['label'])))
```
